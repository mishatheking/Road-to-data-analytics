```python
df.groupby("Base Flavor").agg({"Flavor Rating": ["mean","max","min","count","sum"],"Total Rating": ["mean","max","min","count","sum"]})
```

## Merging DataFrames

```python
df1 = pd.read_csv(r"C:\Users\Misha\Documents\6. Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\Misha\Documents\6. Pandas\LOTR 2.csv")
df2
```

```python
# left_df.merge(right_df)
df1.merge(df2)
```

```python
# Default merge (inner join) on the first column, can be changed using the 'how'  and 'on' parameters
df1.merge(df2, how="inner", on= "FellowshipID" )
```

```python
# merge column can be a list too
df1.merge(df2, how="inner", on= ["FellowshipID", "FirstName"])
```

```python
# specifying distinguishers for shared column
df1.merge(df2, how="inner", on= "FellowshipID", suffixes=('_a', '_b'))
```

```python
# Outter join
df1.merge(df2, how= "outer")
```

```python
# left join
df1.merge(df2, how= "left")
```

```python
# Right join
df1.merge(df2, how= "right")
```

```python
# cross join
df1.merge(df2, how= "cross")
```

## Using Join
A bit more coplicated but better for working with index

```python
df1
```

```python
df2
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
# trying join with index 
df3 = df1.set_index("FellowshipID").join(df2.set_index("FellowshipID"), how = 'cross', lsuffix = '_left', rsuffix = '_right')
df3
```

## Concatnation

Unlike the merge and join that put dataframes beside each other, concatnation put them on top of each other

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join ="outer", axis = 1)
```
